```python
import numpy as np
```

```python
alphas_bipartite = np.load('./alphas_mc_nodes_5_bipartite.npy')
alphas_hyper = np.load('./alphas_mc_nodes_5_hyper.npy')
alphas_fully = np.load('./alphas_mc_nodes_5_fully.npy')
alphas_cir = np.load('./alphas_mc_nodes_5_cir.npy')
alphas_star = np.load('./alphas_mc_nodes_5_star.npy')
alphas_cent = np.load('./alphas_mc_nodes_5_cent.npy')
alphas_dis = np.load('./alphas_MNIST_disconnect_8nodes.npy')
```

```python
nodes = 8
j = 0
lr_list = [0.0001,
0.001,
0.01,
0.015,
0.02,
0.025,
0.03,
0.04,
0.045,
0.05,
0.06,
0.075]
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(10,7.5))
for alpha in alphas_fully:
    plt.plot(lr_list ,alpha[:,j], 'b-', linewidth=3)
for alpha in alphas_cir:
    plt.plot(lr_list ,alpha[:,j], 'r-', linewidth=3)
for alpha in alphas_star:
    plt.plot(lr_list ,alpha[:,j], 'm-', linewidth=3)
for alpha in alphas_hyper:
    plt.plot(lr_list ,alpha[:,j], 'c-', linewidth=3)
plt.plot(lr_list ,alphas_dis[0][:,j][:12], 'y-', linewidth=3)
plt.plot(lr_list, alphas_cent[:,j], 'g-', linewidth = 3)
plt.xlabel(r'Stepsize $\eta$',fontsize=25)
plt.ylabel(r'Tail index $\alpha$',fontsize=25)
plt.tick_params(labelsize=25)
legendList = ['Complete', 'Cycle', 'Star','HyperCube',  'Dis-SGD', 'C-SGD']
plt.legend(legendList, loc='best', fontsize = 23)
# bbox_to_anchor=(1,0.5)
# plt.legend(legendList,bbox_to_anchor=(0,1.02,1,0.2), loc="lower left", borderaxespad=0,fontsize=30,ncol=5)
plt.show()
```
