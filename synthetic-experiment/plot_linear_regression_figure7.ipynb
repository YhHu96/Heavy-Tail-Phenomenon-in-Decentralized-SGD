```python
import numpy as np
# fully
alpha_nodes_fully = np.load('./final_stepsize/alpha_batch_fully.npy')
alpha_nodes_cir = np.load('./final_stepsize/alpha_nodes_cir.npy')
alpha_nodes_hyper = np.load('./final_stepsize/alpha_nodes_hyper.npy')
alpha_nodes_bipartite = np.load('./final_stepsize/alpha_nodes_bipartite.npy')
alpha_nodes_star = np.load('./final_stepsize/alpha_nodes_star_new.npy')
alpha_cent_all = np.load('./final_stepsize/alpha_nodes_cent.npy')
```

```python
nodes = 8
lr_list = np.linspace(0.165,0.205,num = 30)
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(10,7.5))
plt.plot(lr_list, np.median(alpha_nodes_fully, axis = 0), linewidth = 3)
plt.plot(lr_list, np.median(alpha_nodes_cir, axis = 0), linewidth = 3)
plt.plot(lr_list, np.median(alpha_nodes_star, axis = 0), linewidth = 3)
plt.plot(lr_list, np.median(alpha_nodes_hyper, axis = 0), linewidth = 3)
plt.plot(lr_list, np.median(alpha_nodes_bipartite, axis = 0), linewidth = 3)
plt.plot(lr_list, alpha_cent_all[0], linewidth = 3)
plt.xlabel(r'Stepsize $\eta$',fontsize=25)
plt.ylabel(r'Tail index $\alpha$',fontsize=25)
plt.tick_params(labelsize=25)
# bbox_to_anchor=(1.04,0.5)
legendList = ['Complete','Cycle','Star','HyperCube', 'Bipartite', 'C-SGD']
plt.legend(legendList, loc='best', fontsize=23)
plt.show()
```
